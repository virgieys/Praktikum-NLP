```python
# Import library yang dibutuhkan
import pandas as pd
import re
from faker import Faker
```

```python
# Membuat objek Faker yang akan digunakan utk menghasilkan data palsu.
fake = Faker()

# Jumlah data yg diinginkan
jumlah_data = 100

# Inisialisasi list untuk menyimpan data palsu.
data_dummy = []
```

```python
# Fungsi untuk menghapus semua karakter non-digit dari teks.
def remove_punctuation(text):
    return re.sub(r'[^\d]+', '', text)

# Loop sebanyak jumlah_data (variabel yang seharusnya telah didefinisikan sebelumnya).
for _ in range(jumlah_data):
    # Menghasilkan data palsu menggunakan modul Faker.
    nama = fake.name()
    email = fake.email()
    alamat = fake.address()
    nomor = fake.phone_number()

    # Menggunakan fungsi remove_punctuation untuk menghapus karakter non-digit dari nomor telepon.
    nomor = remove_punctuation(nomor)

    # Membuat string data yang berisi informasi palsu.
    data = f"Nama: {nama}\nAlamat Email: {email}\nNomor Telepon: {nomor}\nAlamat Rumah: {alamat}\n"

    # Menambahkan string data ke dalam list data_dummy.
    data_dummy.append(data)

# Menyimpan data palsu ke dalam file dengan nama file_name.
file_name = 'data_dummy.txt'
with open(file_name, 'w') as file:
    for data in data_dummy:
        file.write(data + "\n")

# Pola regular expression untuk mencocokkan alamat email dan nomor telepon.
email_pattern = r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,7}\b'
phone_pattern = r'\d{9,17}'

# Membaca isi file dan menyimpannya dalam variabel data.
with open(file_name, 'r') as file:
    data = file.read()

# Mencocokkan pola email dan nomor telepon dalam data menggunakan re.findall.
email_matches = re.findall(email_pattern, data)
phone_matches = re.findall(phone_pattern, data)
```

```python
# Konversi list data_dummy menjadi DataFrame
df = pd.DataFrame(data_dummy, columns=['Data Dummy'])

# Tampilkan DataFrame
display(df)
```

```python
# Membuat kamus (dictionary) yang berisi pasangan kunci-nilai ('Alamat Email', email_matches) dan ('Nomor Telepon', phone_matches).
data_dict = {'Alamat Email': email_matches, 'Nomor Telepon': phone_matches}

# Membuat DataFrame, dimana kolom 'Alamat Email' berisi email_matches dan kolom 'Nomor Telepon' berisi phone_matches.
data_re = pd.DataFrame(data_dict)
```
