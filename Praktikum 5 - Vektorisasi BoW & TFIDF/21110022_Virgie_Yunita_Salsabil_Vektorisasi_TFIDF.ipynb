### 1. load dataset yang diberikan menjadi dataframe

```python
# Input & tampilkan dataset
file = pd.read_csv('data.csv')
display(file)
```

```python
# Drop kolom
teks = file.drop(['Unnamed: 0', 'userName', 'score'], axis=1)
teks
```

### 2. ubah kolom "content" menjadi vector menggunakan TFIDFVectorizer dan tampilkan hasilnya

```python
# Ambil 5 data dan ubah kolom "content" menjadi vektor TF-IDF
sample = teks.tail(5)  # Ambil 5 data terakhir
teks = sample["content"]
```

```python
# Tampilkan 5 data terakhir
display(teks)
```

```python
# Menampilkan keseluruhan teks
array_teks = np.array(teks)
array_teks
```

- Proses TF-IDF

```python
# Inisialisasi TF-IDF Vectorizer
tfidf = TfidfVectorizer()

# Transformasi teks menjadi vektor TF-IDF
tfidf_matrix = tfidf.fit_transform(teks)
```

```python
# Tampilkan hasil
print("Hasil TF-IDF Vectorizer untuk 5 data terakhir:")
print(tfidf_matrix)
```

### 3. Tampilkan nama fiturnya mengunakan *get_features_name_out*

```python
# Menggunakan metode get_feature_names_out() dari objek tfidf
# utk mendapatkan daftar nama fitur yg digunakan dlm vektor TF-IDF
feature_names = tfidf.get_feature_names_out()

# Menampilkan daftar nama fitur
print("Nama fitur:\n")
print(feature_names)
```

```python
# Loop melalui indeks dan data pada DataFrame 'sample_data'
for i, data_idx in enumerate(sample.index):
    # Menampilkan informasi header untuk setiap data
    print(f"---------------")
    print(f"\nDATA KE-{i + 1}: \n")
    print(f"---------------")

    # Mendapatkan vektor TF-IDF untuk data tertentu dari matriks TF-IDF
    data_tf = tfidf_matrix[i].toarray()

    # Loop melalui indeks dan nama fitur untuk menampilkan nilai TF-IDF
    for j, feature_name in enumerate(feature_names):
        # Menampilkan nilai TF-IDF untuk setiap fitur dalam data tertentu
        print(f"{feature_name}: {data_tf[0][j]:.4f}")
```

```python
# Buat dataframe dari matriks TF-IDF dengan nama kolom sesuai dengan fitur kata
tfidf_df = pd.DataFrame(data=tfidf_matrix.toarray(), columns=feature_names)

# Tampilkan matriks kata
print(tfidf_df)
```

### **Representasi output TFIDF**



- Nilai yang tinggi menunjukkan kata-kata yang relatif penting atau unik untuk data tersebut.
- Nilai yang rendah menunjukkan kata-kata yang muncul secara umum di seluruh korpus.
- Nilai nol menunjukkan kata-kata yang tidak muncul dalam data tersebut.

Sebagai contoh:
- Pada data ke-1, kata "apa" memiliki bobot 0.154998, sedangkan pada data ke-2, kata "apa" memiliki bobot 0.00 (tidak muncul).
- Pada data ke-3, kata "bagus" memiliki bobot 0.311236, menunjukkan kata tersebut relatif penting dalam data ke-3.
