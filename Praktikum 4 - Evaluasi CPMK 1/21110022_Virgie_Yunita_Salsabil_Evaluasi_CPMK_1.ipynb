### ***2. CPMK1- Sub CPMK1.1. `( bobot : 25 )`***


- Lakukan pembersihan meliputi penghapusan punctuation, angka dan karakter yang tidak penting menggunakan menggunakan regex.
- Simpan hasilnya menjadi file csv/excel.


```python
# Membersihkan kolom full_text
def clean_text(text):
    # Menghapus semua karakter selain huruf (a-z, A-Z) dan spasi dari teks.
    cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
    return cleaned_text

# Mengaplikasikan fungsi clean_text ke kolom full_text dan menambahkan hasilnya sebagai kolom baru 'cleaned_text'
subset_data['cleaned_text'] = subset_data['full_text'].apply(clean_text)

# Tampilkan dataset
subset_data
```

```python
# Membersihkan kolom username
def clean_username(username):
    # Menghapus semua karakter selain huruf (a-z, A-Z) dan spasi dari username.
    cleaned_username = re.sub(r'[^a-zA-Z\s]', '', username)
    return cleaned_username

# Mengaplikasikan fungsi clean_username ke kolom username dan menambahkan hasilnya sebagai kolom baru 'cleaned_username'
subset_data['cleaned_username'] = subset_data['username'].apply(clean_username)

# Menampilkan DataFrame yang sudah dimodifikasi
subset_data
```

```python
# Membuat DataFrame baru tanpa kolom 'full_text' dan 'username'
CLEAN = subset_data.drop(['full_text', 'username'], axis=1)

# Menampilkan DataFrame baru yang telah dibuat
CLEAN
```

```python
# Menyimpan DataFrame CLEAN ke dalam file CSV dengan nama '2.Data_Cleaned.csv' tanpa menyertakan indeks
CLEAN.to_csv('2.Data_Cleaned.csv', index=False)
```

### b. Carilah kata-kata slangword yang ada dalam dataset Anda, dengan cara mencocokan dengan kamus KBBI (terlampir).

- Simpan hasilnya dalam bentuk csv/excel.
-  Tampilkan 100 kata slang yang Anda dapatkan dan tampilkan dalam bentuk Gunakan `(bobot:25)`

```python
# Membaca file CSV yang berisi kata-kata slang
slang = pd.read_csv('stopwords.csv')

# Tampilkan data
slang
```

```python
# Mengambil daftar kata-kata slang dari kolom 'Slangwords' dalam DataFrame 'slang' dan mengonversikannya menjadi list
slang_words = slang['Slangwords'].tolist()

# Fungsi untuk mencari kata-kata slang dalam teks
def find_slang(text):
    # Mengambil token (kata-kata) dari teks
    words = text

    # Mencari kata-kata slang yang ada dalam teks
    found_slang = [word for word in words if word in slang_words]

    # Menggabungkan kata-kata slang yang ditemukan menjadi satu string, dipisahkan oleh koma
    return ', '.join(found_slang)

# Menerapkan fungsi find_slang ke kolom 'parse_text' dalam DataFrame 'data_parsing' dan menambahkan hasilnya sebagai kolom baru 'found_slang_in_text'
data_parsing['found_slang_in_text'] = data_parsing['parse_text'].apply(find_slang)

# Menampilkan DataFrame 'data_parsing' yang telah dimodifikasi dengan tambahan kolom 'found_slang_in_text'
data_parsing
```

```python
# Mengambil kata-kata slang yang telah ditemukan
found_slang_list = data_parsing['found_slang_in_text'].str.split(', ').explode().tolist()

# Menyimpan seluruh slang ke dalam 'all slang'
all_slang = found_slang_list
```

```python
# Mengambil 100 kata terakhir
slang_100 = found_slang_list[-100:]

# Menampilkan 100 kata slang terakhir
slang_100
```

```python
# Membuat DataFrame dari list kata-kata slang
all_slang_df = pd.DataFrame({'all_slang': all_slang})
all_slang_df
```

```python
# Simpan DataFrame ke dalam file CSV
all_slang_df.to_csv('3B.Data_Slang.csv', index=False)
```
